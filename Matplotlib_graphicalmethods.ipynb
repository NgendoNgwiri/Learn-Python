```python
#Solving cubic functions
import numpy as np

import matplotlib.pyplot as plt

x= np.arange(-10.0,10.0,1)


#y= (-x**3) + (2(x**2)) + (4*x)

y= -(x**3) +(3*x)+ 2


plt.plot(x,y,'r--')

plt.show()
```
